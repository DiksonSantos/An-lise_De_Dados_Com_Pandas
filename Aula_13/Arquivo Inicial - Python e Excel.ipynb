# Integração Python + Excel

### 2 formas:

1. Pandas (Melhor para arquivos Extensos)
    - Mais usada no geral
    - Trata o Excel como uma base de dados (NÃO ENXERGA AS FORMULAS & GRÁFICOS CONTIDOS NA PLANILHA...)
    - Faz o que quiser com o arquivo
    - Pode desfazer a estrutura original do arquivo, caso queira editar
    
2. Openpyxl (Melhor para tratar especificidades)
    - Trata o Excel como uma planilha mesmo que pode ter várias coisas (CONSERVA FORMULAS E Porém pode perder os GRÁFICOS DAS PLANILHAS)
    - Edita "como se fosse um VBA"
    - Menos eficiente
    - Mantém mais a estrutura original do arquivo, mas cuidado porque não necessariamente tudo, então tem que testar

### Desafio

- Temos uma planilha de produtos e serviços. Com o aumento de imposto sobre os serviços, temos que atualizar o preço dos produtos impactados pela mudança.

Novo Multiplicador Imposto: 1.5

```python
# USANDO PANDAS:
import pandas as pd
tabela = pd.read_excel('Produtos.xlsx')
display(tabela)
```

```python
# atualizando a cotação e as colunas seguintes
#->Localizar local p/ Inserção da nova formula p/ novo Calculo de imposto.
#-> tabela.loc[linha, coluna]

#-> Localizar todas as linha 'Serviços' da Coluna 'Tipo'
# Coluna Multiplicador Recebera o valor 1.5 se a linha em 'Tipo' for == 'Serviço'
# Onde na coluna Tipo a linha for == Serviço, a linha referente em
#.. Nultiplicador recebe valor 1.5

tabela.loc[tabela["Tipo"]=="Serviço", "Multiplicador Imposto"] = 1.5


# As malditas aspas simples Fazem a linha de codigo NÃO funcionar. 
# tabela.loc[tabela['Tipo']=='Serviços','Multiplicador Imposto'] = 1.5




display(tabela['Multiplicador Imposto'])
```

```python

#Coluna 'Preço Base Reais' recebe novo calculo:
#tabela['Preço Base Reais'] = tabela['Multiplicador Imposto'] * tabela['Preço Base Original']
tabela["Preço Base Reais"] = tabela["Multiplicador Imposto"] * tabela["Preço Base Original"]

# Se der o mesmo nome do xlsx original; ele Salva e Substituindo o Arquivo .xlsx Original:
# tabela.to_excel('Produtos.xlsx')
# A Linha 'index=False' Faz com que não exista uma Coluna 0 Com uma numero por linha:
tabela.to_excel("Produtos_Made_In_Pandas.xlsx", index=False)
#tabela.to_excel("Produtos_Made_In_Pandas.xlsx")


tabela2 = pd.read_excel('Produtos.xlsx')
display(tabela2)
```

```python
# openpyxl
from openpyxl import Workbook, load_workbook

planilha_ = load_workbook("Produtos_Made_In_Pandas.xlsx")

aba_ativa = planilha_.active

# Percorre todas as celulas da coluna 'C' que tiverem valores
for celula in aba_ativa["C"]:
    if celula.value == 'Serviço':
        # A variavel recebe o valor/linha 'Serviço'
        linha = celula.row
        aba_ativa[f'D{linha}'] = 1.5
        
planilha_.save("MadeIn_WorkBook.xlsx")



#_______________________________________________________
tabela3 = pd.read_excel("MadeIn_WorkBook.xlsx")
display(tabela3)
```
